```python
from transformers import pipeline, AutoModelForSequenceClassification, AutoTokenizer

model = AutoModelForSequenceClassification.from_pretrained("saved_model")
tokenizer = AutoTokenizer.from_pretrained("saved_model")

classifier = pipeline(
    "sentiment-analysis",
    model=model,
    tokenizer=tokenizer,
    return_all_scores=True,
    device=0  # use GPU again
)
```

```python


batch_size = 64  # safe for Colab GPU
texts = df["sentences"].tolist()

results = []
for i in range(0, len(texts), batch_size):
    batch = texts[i:i+batch_size]
    results.extend(classifier(batch))
```

```python
def extract_sentiment(scores):
    # 'scores' is a list of dictionaries, e.g., [{'label': 'LABEL_0', 'score': 0.006}, ...]
    # Find the dictionary with the highest score
    best = max(scores, key=lambda x: x["score"])
    return best["label"], round(best["score"], 3)

df[["sentiment", "sentiment_score"]] = [
    extract_sentiment(r) for r in results
]
```

```python
df.tail(50)
```

```python
df[['sentences', 'sentiment_score']].tail(60)
```

```python
def categorize_sentiment(score,sentiment):
    if score >= 0.48:
        return "Positive"
    elif score <= 0.402 or (score<=0.467 and sentiment=="3 stars"):
        return "Negative"
    else:
        return "Neutral"

df['sentiment_label'] = df.apply(lambda row: categorize_sentiment(row['sentiment_score'], row['sentiment']), axis=1)

# Check results
print(df.head())
```
