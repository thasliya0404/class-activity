```python
import pandas as pd
```

```python
data1 = pd.read_csv('/content/population.csv')
data1
```

```python
#Standardize country names across all datasets
data2.rename(columns={'name': 'country'}, inplace=True)
print("Unique countries in data1:")
display(data1['country'].unique())

print("\nUnique countries in data2:")
display(data2['country'].unique())

print("\nUnique countries in data3:")
display(data3['country'].unique())

print("\nUnique countries in data4:")
display(data4['Country'].unique())
```

```python
#Create new columns such as internet penetration rate (internet users / population  100)
merged_data = pd.merge(data1, data3, on = 'country', how='inner')
internet_penetration_rate = (merged_data['internet_users'] / merged_data['population']) * 100
merged_data['internet_penetration_rate'] = internet_penetration_rate
merged_data
```

```python
#Merge the datasets based on the country name
merged_data1 = pd.merge(merged_data, data2, on='country', how='inner')
merged_data2 = pd.merge(merged_data1, data4, left_on='country', right_on='Country', how='inner')
merged_data2.drop('Country', axis=1, inplace=True)
merged_data2
```

```python
# Handle unmatched or missing records
merged_data2.info()
```

```python
#Validate the data for consistency and completeness
missing_values = merged_data2.isnull().sum()
print("Missing values:")
display(missing_values)
duplicated_values = merged_data2.duplicated().sum()
print("Dulicated values:", duplicated_values)
```

```python
#Find countries with highest internet penetration rates
highest_penetration_countries = merged_data.sort_values(by='internet_penetration_rate', ascending=False)
display(highest_penetration_countries[['country', 'internet_penetration_rate']].head(1))
```

```python
#Calculate average literacy rate across countries
average_literacy_rate = merged_data2['literacy_rate'].mean()
print("average literacy rate:", average_literacy_rate)
```

```python
#Investigate relationships between GDP, population, literacy, and internet
correlation_matrix = merged_data2[['population', 'literacy_rate', 'internet_users', 'GDP']].corr()
print(correlation_matrix)
```

```python
import sqlite3
conn = sqlite3.connect('country.db')
cursor = conn.cursor()
```

```python
merged_data2.to_sql('Final_database',conn,if_exists='replace',index=True)
```

```python
# a. Countries with highest internet penetration
cursor.execute("""SELECT country, internet_penetration_rate
FROM Final_database ORDER BY internet_penetration_rate DESC LIMIT 1;""")
print(cursor.fetchone())
```

```python
# b. Average literacy rate
cursor.execute("""SELECT AVG(literacy_rate)
FROM Final_database;""")
print(cursor.fetchone())
```

```python
#Correlation between literacy rate and internet penetration
cursor.execute("""SELECT
(COUNT(*) * SUM(literacy_rate * internet_penetration_rate) - SUM(literacy_rate) * SUM(internet_penetration_rate)) /
(SQRT(COUNT(*) * SUM(literacy_rate*literacy_rate) - SUM(literacy_rate) * SUM(literacy_rate)) * SQRT(COUNT(*) * SUM(internet_penetration_rate * internet_penetration_rate) -
SUM(internet_penetration_rate) * SUM(internet_penetration_rate))) AS correlation FROM Final_database;""")
print(cursor.fetchall())
```

```python
# GDP per capita analysis using cursor.execute
cursor.execute("""SELECT country, (GDP / population) AS GDP_per_capita
FROM Final_database;
""")
print(cursor.fetchall())
```

```python
#Which countries have a GDP per capita above $10,000?
cursor.execute("""SELECT country, (GDP / population) AS GDP_per_capita
FROM Final_database
ORDER BY GDP_per_capita DESC;""")
print(cursor.fetchall())
```

```python
#What is the total population covered in the dataset?
cursor.execute("""SELECT SUM(population)
FROM Final_database;""")
print(cursor.fetchone())
```

```python
#Which countries have the lowest literacy rates, and how does that impact internet access?
cursor.execute( """SELECT country, literacy_rate, internet_penetration_rate
FROM Final_database
ORDER BY literacy_rate LIMIT 1;""")

print(cursor.fetchall())
```

```python
#What are the top 5 wealthiest countries by total GDP, and how does that compare with population size?
cursor.execute("""SELECT country, GDP, population
FROM Final_database
ORDER BY GDP DESC
LIMIT 5;""")
print(cursor.fetchall())
```

```python
#Find countries where internet users exceed 70% of the population
cursor.execute("""SELECT country, internet_penetration_rate
FROM Final_database
WHERE internet_penetration_rate > 70;""")
print(cursor.fetchall())
```

```python
#What is the average GDP per capita for countries with internet penetration above 50%?
cursor.execute("""SELECT AVG(GDP / population) AS average_gdp_per_capita
FROM Final_database
WHERE internet_penetration_rate > 50;""")
print(cursor.fetchone())
```

```python
#How many countries have a literacy rate above 90%, and what is their average internet penetration?
cursor.execute("""SELECT COUNT(country) AS num_countries, AVG(internet_penetration_rate) AS average_internet_penetration
FROM Final_database
WHERE literacy_rate > 90;""")
print(cursor.fetchone())
```
