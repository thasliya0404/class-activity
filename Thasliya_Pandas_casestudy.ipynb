```python
url='https://raw.githubusercontent.com/laxmimerit/All-CSV-ML-Data-Files-Download/master/IMDB-Movie-Data.csv'
#1) Import pandas, numpy
import pandas as pd
import numpy as np
#2) Load dataset using pd.read_csv()
df = pd.read_csv(url)
df
```

```python
#3) Display: shape, head, column names, datatypes, missing value
print('shape',df.shape)
print('\nhead\n',df.head())
print('\ncolumns\n',df.columns)
print('\ndata types\n',df.dtypes)
print('\nmissing values\n',df.isnull().sum())
```

```python
#4) Check and remove duplicates
print('duplicates',df.duplicated().sum())
```

```python
#5) Rename columns ("Title" > "Movie_Title")
df = df.rename(columns = {'Title':'Movie_Title'})
df
```

```python
#6) Get unique genres
df['Genre'].unique()
```

```python
#7) Extract number of genres per movie
df['Genre'].apply(lambda x:len(x.split(',')))
```

```python
#8) Handle missing values in 'Revenue (Millions)' and 'Metascore
x = df['Revenue (Millions)'].fillna(df['Revenue (Millions)'].median())
y = df['Metascore'].fillna(df['Metascore'].median())
print('missing values in Revenue (Millions)', x)
print('missing values in Metascore', y)
```

```python
#9) Top 5 movies by Rating
top5_rating = df.sort_values(by = 'Rating',ascending=False).head()
top5_rating[['Movie_Title','Rating']]
```

```python
#10) Movies with highest/lowest revenue
highest_rev = df.sort_values(by = 'Revenue (Millions)',ascending = False).head()
lowest_rev = df.sort_values(by = 'Revenue (Millions)',ascending = True).head()
print(highest_rev[['Movie_Title','Revenue (Millions)']])
print(lowest_rev[['Movie_Title','Revenue (Millions)']])
```

```python
#11) Avg. rating by Director
avg_rating_director = df.groupby('Director')['Rating'].mean()
avg_rating_director
```

```python
#12) Movies per year
movies_per_year = df['Year'].value_counts().sort_index()
movies_per_year
```

```python
#13) Movies with rating > 8 & revenue > 100M
x = df[(df['Rating']>8) & (df['Revenue (Millions)']>100)]
x[['Movie_Title','Rating','Revenue (Millions)']]
```

```python
#14) Group by Director: avg Rating, Revenue
z = df.groupby('Director')[['Rating','Revenue (Millions)']].mean()
z
```

```python
#15) Group by Year: count, avg Rating
x = df.groupby('Year').agg(movie_count = ('Movie_Title','count'),avg_rating = ('Rating','mean'))
x
```

```python
#16) Create a new column Profitability:
#a) Revenue divided by Duration
#b) Find top 10 most profitable movies
#a
df['Profitability'] = df['Revenue (Millions)'] / df['Runtime (Minutes)']
print(df['Profitability'] )
#b
top10_profitable = df.sort_values(by = 'Profitability', ascending = False).head(10)
print(top10_profitable[['Movie_Title', 'Profitability']])
```

```python
#17) Create a column that categorizes movies by rating:
#a) Rating < 6: Poor
#b) 6 ≤ Rating < 7.5: Average
#c) Rating ≥ 7.5: Excellent

def rating_category(r):
    if r < 6:
        return 'Poor'
    elif 6 <= r < 7.5:
        return 'Average'
    else:
        return 'Excellent'

df['Rating_Category'] = df['Rating'].apply(rating_category)
print(df['Rating_Category'] )
```

```python
#18) Create a custom function to score movies based on:
#i) Apply row['Rating'] * 10 + row['Revenue (Millions)'] as High score
#Apply a function to dataset to assign the score
def movie_score(row):
    return row['Rating']*10 + row['Revenue (Millions)']

df['High_Score'] = df.apply(movie_score, axis=1)
print(df['High_Score'])
```
